```python
import matplotlib.pyplot as plt
import numpy as np

eeg_array = np.array(eeg_data)

# Select a subset of EEG channels (e.g., first 5)
num_channels_to_plot = 5
plt.figure(figsize=(12, 6))

for i in range(num_channels_to_plot):
    plt.plot(timestamps, eeg_array[:, i], label=f'Channel {i+1}')

plt.xlabel("Time (s)")
plt.ylabel("EEG Amplitude")
plt.title("EEG Signal for First 5 Channels")
plt.legend()
plt.show()
```

```python
import pandas as pd

# Convert EEG data into a DataFrame for better analysis
df_eeg = pd.DataFrame(eeg_array)
df_eeg['timestamp'] = timestamps  # Add timestamps column

# Display basic statistics
print(df_eeg.describe())
```

```python
# Find channels with extreme values
threshold = 900  # Set based on your dataset (typically 100-200 µV for EEG)
artifact_channels = (df_eeg.drop(columns=['timestamp']).abs() > threshold).sum()

# Print channels with excessive spikes
print("Possible artifact channels:")
print(artifact_channels[artifact_channels > 50])  # More than 50 extreme values
```

```python
from scipy.signal import butter, filtfilt

# Define bandpass filter
def bandpass_filter(data, lowcut=1.0, highcut=50.0, fs=120, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data, axis=0)

# Apply the filter
filtered_eeg = bandpass_filter(eeg_array)

# Convert back to DataFrame
df_filtered_eeg = pd.DataFrame(filtered_eeg)
df_filtered_eeg['timestamp'] = timestamps

# Plot before & after filtering for one channel
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(timestamps[:1000], eeg_array[:1000, 0], label="Raw EEG (Noisy)", alpha=0.7)
plt.plot(timestamps[:1000], filtered_eeg[:1000, 0], label="Filtered EEG (1-50Hz)", alpha=0.9)
plt.xlabel("Time (s)")
plt.ylabel("EEG Amplitude")
plt.title("EEG Signal Before & After Filtering (First Channel)")
plt.legend()
plt.show()
```

```python
import numpy as np

# Compute standard deviation per channel
channel_std = df_filtered_eeg.drop(columns=['timestamp']).std()

# Set a threshold (adjust based on your data)
std_threshold = channel_std.mean() + 2 * channel_std.std()

# Identify bad channels
bad_channels = channel_std[channel_std > std_threshold].index.tolist()

print(f"Bad EEG channels: {bad_channels}")
```

```python
# Drop bad channel
df_cleaned_eeg = df_filtered_eeg.drop(columns=[18])

print(f"Remaining channels after cleaning: {df_cleaned_eeg.shape[1] - 1}")  # Exclude timestamp column
```

```python
# Extract event marker stream
marker_stream = next(s for s in streams if s['info']['name'][0] == 'MyMarkerStream3')

# Get timestamps and event labels
marker_timestamps = np.array(marker_stream['time_stamps'])
marker_labels = np.array(marker_stream['time_series'])

print(f"Total Markers: {len(marker_timestamps)}")
print("First 10 Marker Timestamps & Labels:")
for i in range(min(10, len(marker_timestamps))):
    print(f"Time: {marker_timestamps[i]:.2f}  |  Event: {marker_labels[i]}")
```

```python
# Find closest EEG timestamp for each marker
aligned_indices = [np.argmin(np.abs(timestamps - t)) for t in marker_timestamps]

# Create a new column in the EEG DataFrame for event markers
df_cleaned_eeg['event'] = ''
for i, idx in enumerate(aligned_indices):
    df_cleaned_eeg.at[idx, 'event'] = marker_labels[i][0]  # Assuming marker labels are stored as lists
```

```python
# Extract Eye-Tracking stream
eye_stream = next(s for s in streams if s['info']['name'][0] == 'Tobii')

# Get timestamps & gaze data
eye_timestamps = np.array(eye_stream['time_stamps'])
eye_data = np.array(eye_stream['time_series'])

print(f"Total Eye-Tracking Samples: {eye_data.shape[0]}")
print(f"Eye-Tracking Data Shape: {eye_data.shape}")  # (samples, features)
print("First 5 Eye-Tracking Samples:\n", eye_data[:5])
```

```python
# Find closest EEG timestamp for each eye-tracking timestamp
aligned_eye_indices = [np.argmin(np.abs(timestamps - t)) for t in eye_timestamps]

# Convert eye-tracking data into a DataFrame
df_eye_tracking = pd.DataFrame(eye_data, columns=[f'eye_feature_{i}' for i in range(eye_data.shape[1])])
df_eye_tracking['timestamp'] = eye_timestamps

# Merge EEG and Eye-Tracking on timestamps
df_merged = pd.merge_asof(df_cleaned_eeg, df_eye_tracking, on="timestamp", direction="nearest")

print(f"Final Merged Data Shape: {df_merged.shape}")
print(df_merged.head())
```

```python
import matplotlib.pyplot as plt

# Select one EEG channel and one gaze feature to plot
eeg_channel = df_merged.columns[2]  # Change index to select another channel
eye_feature = df_merged.columns[-2]  # Second last column (eye-tracking feature)

plt.figure(figsize=(12, 6))

# Plot EEG
plt.subplot(2, 1, 1)
plt.plot(df_merged['timestamp'][:1000], df_merged[eeg_channel][:1000], label=f'EEG - {eeg_channel}', color='blue', alpha=0.7)
plt.ylabel("EEG Amplitude")
plt.title(f"EEG Signal ({eeg_channel}) Over Time")
plt.legend()

# Plot Eye-tracking
plt.subplot(2, 1, 2)
plt.plot(df_merged['timestamp'][:1000], df_merged[eye_feature][:1000], label=f'Eye-Tracking - {eye_feature}', color='red', alpha=0.7)
plt.xlabel("Time (s)")
plt.ylabel("Gaze Position / Pupil Size")
plt.title(f"Eye-Tracking Data ({eye_feature}) Over Time")
plt.legend()

plt.tight_layout()
plt.show()
```
